```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.myntra.com/women-sweaters-sweatshirts")
```

```python
html=driver.page_source
```

```python
from bs4 import BeautifulSoup
import pandas as pd
soup = BeautifulSoup(html, "html.parser")
```

```python

soup.find("li", {"class": "product-base"})
```

```python
df_content = []
for product in soup.find_all("li", {"class": "product-base"}):
    product_brand = product.find("h3", {"class": "product-brand"}).text
    product_product = product.find("h4", {"class": "product-product"}).text
    price = product.find("span", {"class": "product-discountedPrice"}) or product.find("div", {"class": "product-price"})
    price = price.text if price else "no price"
    discount = product.find("span", {"class": "product-discountPercentage"})
    discount = discount.text if discount else "no discount"
    orignal_price = product.find("span", {"class": "product-strike"})
    orignal_price = orignal_price.text if orignal_price else "no orignal"

    df_row =[product_brand, product_product, price, discount, orignal_price]
    df_content.append(df_row)

    

df_headers=['product_brand', 'product_product', 'price', 'discount', 'orignal_price']
df= pd.DataFrame(df_content, columns=df_headers)
df
```

```python
df.to_csv("freddie_sweater.csv", index=False)
```
