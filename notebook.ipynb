```python
from selenium import webdriver
from bs4 import BeautifulSoup 
import pandas as pd 
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
driver.get("https://www.amazon.in/b/?_encoding=UTF8&node=1318164031&bbn=976389031&ref_=Oct_d_odnav_d_1318157031_7&pd_rd_w=fhiwL&content-id=amzn1.sym.245a683c-9fb0-42b8-84a1-4ad3266c7763&pf_rd_p=245a683c-9fb0-42b8-84a1-4ad3266c7763&pf_rd_r=Z9YZAZKF083G3KHQ4VYX&pd_rd_wg=HtYIx&pd_rd_r=d70b60af-bb91-4585-ad10-c66d7f9f0e4d")
html=driver.page_source
soup = BeautifulSoup(html, "html.parser")
```

```python
soup.find("span", {"class": "a-list-item"})
```

```python
df_content = []
for amazon in soup.find_all("li", {"class": "octopus-pc-item octopus-pc-item-v3"}):
     price = amazon.find("span", {"class": "a-price-whole"}).text.rstrip('.')
     name = amazon.find("span", {"class": "a-size-base a-color-base"}).text.strip()
     star = amazon.find("span", {"class": "a-icon-alt"})
     star = star.text[0:3] if star else "no rating"
     orignal_price = amazon.find("span", {"class": "a-size-mini"}) or amazon.find("span", {"class": "a-color-tertiary"}) or amazon.find("span", {"class": "a-text-strike"})
     orignal_price = orignal_price.text.strip() if orignal_price else "no price"
     people_rating = amazon.find("span", {"class": "a-size-mini a-color-tertiary"})
     people_rating = people_rating.text if people_rating else "no rating count"
     df_row =[price, name, star, orignal_price, people_rating]
     df_content.append(df_row)
df_headers=['price', 'name', 'star', 'orignal_price', 'people_rating']
df= pd.DataFrame(df_content, columns=df_headers)
df
```

```python

df.to_csv("amazon_books.csv", index=False)
```
