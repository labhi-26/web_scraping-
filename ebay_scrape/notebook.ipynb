```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
dfs = []
counter = 1
while counter < 1001:
    session = requests.Session()
    r = session.get(f"https://books.toscrape.com/catalogue/page-{counter}.html")
    counter = counter + 1
    if r.ok:
        page_source = r.text
        soup = BeautifulSoup(page_source, "html.parser")
        df_info = []
        for product in soup.find_all("article", {"class": "product_pod"}):
            title = product.find("h3")
            title = title.find("a")['title']
            price = product.find("p", {"class": "price_color"}).text[1:]
            rating = product.find("p", {"class": "star-rating"})
            rating = rating['class'][1]
            in_stock = product.find("p", {"class": "instock availability"}).text.strip()
            df_row = [title, price, rating, in_stock]
            df_info.append(df_row)

        df_headers = ['title', 'price', 'rating', 'in_stock']
        df = pd.DataFrame(df_info, columns=df_headers)
        dfs.append(df)
    else:
        break

final_df = pd.concat(dfs, ignore_index=True)
final_df
```

```python

final_df.to_csv("book_store_info.csv", index=False)
```
